```python
df=pd.read_csv('../input/e-commerce-shoppers-behaviour-understanding/train_data_v2.csv')
df.head(10)
```

**Data pre-processing**

```python
df.isna().sum()
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent', 
                        missing_values=np.nan)
imputer = imputer.fit(df[['Month_SeasonalPurchase','Gender','CustomerType','Cookies Setting','Education','Marital Status']])
df[['Month_SeasonalPurchase','Gender','CustomerType','Cookies Setting','Education','Marital Status']] = imputer.transform(df[['Month_SeasonalPurchase','Gender','CustomerType','Cookies Setting','Education','Marital Status']])
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
numeric_columns = df.select_dtypes(include='number').columns
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
df[numeric_columns] = pd.DataFrame(imp.fit_transform(df[numeric_columns]), columns=numeric_columns)
```

```python
df.isna().sum()
```

```python
X = df.drop(['Made_Purchase'], axis=1)
y = df['Made_Purchase']
```

```python
y.replace({False: 0, True: 1}, inplace=True)
y
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')

e=enc.fit_transform(X[['Month_SeasonalPurchase','CustomerType','Gender','Cookies Setting','Education','Marital Status']])
enc_df = pd.DataFrame(e.toarray())
# merge with main X_train

X = X.join(enc_df,lsuffix="l")
X
```

```python
X.shape
```

```python
X = X.drop(['Month_SeasonalPurchase','CustomerType','Gender','Cookies Setting','Education','Marital Status'], axis=1)
```

```python
X.shape
```

```python
X.columns = X.columns.astype(str)
```

```python
X_train=X
y_train=y
```

```python
X_train
```
